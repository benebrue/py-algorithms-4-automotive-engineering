```python
# This if else is a fix to make the file available for Jupyter and Travis CI
import os


def find_mat(filename):
    if os.path.isfile(filename):
        file = filename
    else:
        file = '04_mini-projects/' + filename
    return file
```

```python
import scipy.io
```

```python
def load_measurement(filename, test_size):
    dat = scipy.io.loadmat(find_mat(filename))
    return dat["A"][:-test_size], dat["A"][-test_size:], dat["B"][:-test_size], dat["B"][-test_size:]
```

```python
load_measurement("dat1.mat", 200)
```
