```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append("04_mini-projects")

from solution_load_measurement import *
```

```python
X_train, y_train, X_test, y_test = load_measurement('dat1.mat', test_size=400)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X=X_train, y=y_train)
```

```python
from sklearn.neural_network import MLPRegressor

mlp = MLPRegressor(hidden_layer_sizes=(20, 40, 20), activation="relu", max_iter=10000, verbose=True, 
                   random_state=1, tol=1e-6)
mlp.fit(X=X_train, y=y_train[:, 0])
```

```python
plt.figure(figsize=(16, 6))
plt.plot(y_test, 'k')
plt.plot(lin_reg.predict(X_test), 'r')
plt.plot(mlp.predict(X_test), 'g')
plt.legend(['y_test', 'Linear Model', 'Multi-Layer Perceptron'])
```
