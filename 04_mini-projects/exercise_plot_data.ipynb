```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
sys.path.append("04_mini_projects")
from solution_load_measurement import *
```

```python
X_train, y_train, X_test, y_test = load_measurement("dat1.mat", test_size=400)
```

```python
fig, axs = plt.subplots(3)
fig.set_figheight(15)
fig.set_figwidth(15)
fig.suptitle('Plot of the Data')
axs[0].plot(np.arange(0, len(X_train[:, 0] + 1)), X_train[:, 0], color="red")
axs[0].plot(np.arange(len(X_train[:, 0]) + 1, len(X_train[:, 0]) + len(X_test[:, 0]) + 1), X_test[:, 0], color="blue")
axs[0].set_ylabel("velocity")
axs[1].plot(np.arange(0, len(X_train[:, 1] + 1)), X_train[:, 1], color="red")
axs[1].plot(np.arange(len(X_train[:, 1]) + 1, len(X_train[:, 1]) + len(X_test[:, 1]) + 1), X_test[:, 1], color="blue")
axs[1].set_ylabel("acceleration")
axs[2].plot(np.arange(0, len(y_train + 1)), y_train, color="red")
axs[2].plot(np.arange(len(y_train) + 1, len(y_train) + len(y_test) + 1), y_test, color="blue")
axs[2].set_ylabel("power")
plt.show()
```
