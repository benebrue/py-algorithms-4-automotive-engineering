```python
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
liste = [i for i in range(1500)]
liste2 = [random.gauss(0, 1) for i in liste]
```

```python
plt.hist(liste2, bins = 40)
plt.xlabel("Data")
plt.ylabel("Probability density")
plt.title(r"Histogram of $\mathcal{N}(0, 1)$")
plt.show()
```
