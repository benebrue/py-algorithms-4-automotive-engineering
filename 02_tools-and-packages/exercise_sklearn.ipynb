```python
from sklearn import neural_network
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
digits = datasets.load_digits()
```

```python
digits_X = digits.images.reshape((len(digits.images), -1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits_X, digits.target, test_size=0.3, shuffle=True)
```

```python
clf = neural_network.MLPClassifier(hidden_layer_sizes=(5,), verbose=True, max_iter=1000, random_state=1)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
metrics.plot_confusion_matrix(clf, X_test, y_test)
plt.show()
```
