```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([3, -8])
X.shape
```

```python
time = np.linspace(0, 10, 1000)
A = np.vstack((np.sin(time), np.cos(time))).T
A.shape
```

```python
b = A @ X
print(b)
b.shape
```

```python
b = b + np.random.normal(loc = 0, scale = 0.01, size = (1000))
```

```python
print(b)
```

```python
%matplotlib inline
plt.plot(b)
plt.show()
```

```python
X_hat = np.linalg.solve(a=(A.T @ A), b=(A.T @ b))
X_hat
```

```python
plt.plot(b, "k.")
plt.title("A linear regression task")
plt.plot((A @ X_hat).ravel(), "r:")
plt.legend(["Noisy data", "Least squares estimate"])
plt.show()
```
